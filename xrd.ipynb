```python
from pymatgen.ext.matproj import MPRester
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pymatgen.core.structure import Structure
from pymatgen.analysis.diffraction.xrd import XRDCalculator

with MPRester("sFKdeFLD37yMl1L8Kxu1jUecE8tlqVep") as m:
    
    #Téléchargement du fichier CIF
    fich_cif = m.get_data("mp-867193", data_type='cif')
    
    #Ecriture du fichier
    with open("mp-867193.cif","w") as f:
        f.write(fich_cif['cif'])
        
    #Structure de travail
    struct = Structure.from_file("mp-867193.cif")
    #print(structure)
    
    xrd_calc = XRDCalculator(wavelength='CuKa')
    
    pattern = xrd_calc.get_pattern(struct)

    num = ["Premier pic de diffraction :" , "Deuxième pic de diffraction :","Troisième pic de diffraction :"]
    for i in range(3):
        print(num[i])
        print("")
        print("   1) Angle 2Θ = ", pattern.x[i], "[°]")
        print("   2) Intensité relative = ", pattern.y[i], "[%]")
        print("   3) Indices (hkl) et multiplicité: ", pattern.hkls[i])
        print("\n")
        

    # Affichage du diffractogramme
    print("Diffractogramme de Sr2BiAu pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å (se limitant aux 3 premiers pics): ")

    xrd_calc.show_plot(struct, two_theta_range=(0, 32), annotate_peaks=True)
```
